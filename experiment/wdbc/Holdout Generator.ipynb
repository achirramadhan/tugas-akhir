```python
import pandas as pd
import numpy as np
```

```python
dataset_dir = "/root/skripsi/datasets/"
holdout_dir = "holdout/"
```

```python
df = pd.read_csv(dataset_dir + "wdbc/wdbc.data", header=None)
```

```python
pd.set_option('display.max_columns', None)
display(df.head())
```

```python
df.isnull().any()
```

```python
pd.set_option('display.max_columns', None)
df.describe()
```

```python
X = df.drop([0, 1], axis=1).values
y = df[[1]].values.flatten()
```

```python
display(X[:2])
display(y[:2])
```

```python
y = np.array([0 if u == 'B' else 1 for u in y])
```

```python
display(y[:2])
```

```python
n_holdout = 10
```

```python
from sklearn.model_selection import StratifiedKFold

full_df = pd.DataFrame(np.hstack([X, np.expand_dims(y, axis=-1)]))
full_df.to_csv(holdout_dir + "wdbc.csv", index=False)

holdout_skf = StratifiedKFold(n_splits=n_holdout, shuffle=True, random_state=42)
for i, (train_id, test_id) in enumerate(holdout_skf.split(X, y)):
    X_train = X[train_id]
    y_train = y[train_id]
    X_test = X[test_id]
    y_test = y[test_id]
    
    data_train = np.hstack([X_train, np.expand_dims(y_train, axis=-1)])
    data_test = np.hstack([X_test, np.expand_dims(y_test, axis=-1)])
    
    df_train = pd.DataFrame(data_train)
    df_test = pd.DataFrame(data_test)

    df_train.to_csv(holdout_dir + ("wdbc-%d-train.csv" % i), index=False)
    df_test.to_csv(holdout_dir + ("wdbc-%d-test.csv" % i), index=False)
```
