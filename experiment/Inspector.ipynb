```python
import sys
sys.path.insert(1, '../')

import pandas as pd
import numpy as np
from IMLI import IMLI
```

```python
DATASET_NAME = "liver"
```

```python
df_total = pd.read_csv(("%s/holdout/%s.csv" % (DATASET_NAME, DATASET_NAME)))
display(df_total.tail())
```

```python
# df_total[df_total['1'] < 4.5][df_total['2'] >= 1125][df_total['4'] == 1].shape
```

```python
df_total_np = df_total.values
X_total = df_total_np[:, :-1]
y_total = df_total_np[:, -1]
```

```python
features = IMLI().generate_features(
        X_total,
        y_total,
        categorical_features_id=[],
        discretizer="entropy",
#         n_thresholds=9
    )
```

```python
model = IMLI()
model.fit(X_total[:20], y_total[:20], features)
```

```python
new_X, features = model._preprocess(X_total)
```

```python
for f in features:
    print(f)
```
